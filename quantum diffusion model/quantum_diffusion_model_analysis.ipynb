```python
import pickle
import matplotlib.pyplot as plt

def plot_loss_comparison(file1, file2, output_filename='loss_comparison.png'):
    """
    Plots a comparison of losses from two files and saves the graph as a PNG.

    Parameters:
    - file1: The path to the first loss file (e.g., 'epoch_losses40T8x8.pkl').
    - file2: The path to the second loss file (e.g., 'epoch_losses10T8x8.pkl').
    - output_filename: The name of the output file to save the graph (default: 'loss_comparison.png').
    """
    
    # Load losses from the first file
    with open(file1, 'rb') as f:
        losses_1 = pickle.load(f)

    # Load losses from the second file
    with open(file2, 'rb') as f:
        losses_2 = pickle.load(f)

    # Create the plot
    plt.figure(figsize=(12, 6))

    # Plot losses for file1
    plt.plot(range(1, len(losses_1) + 1), losses_1, marker='o', linestyle='-', label='File 1', markersize=8)

    # Plot losses for file2
    plt.plot(range(1, len(losses_2) + 1), losses_2, marker='s', linestyle='-', label='File 2', markersize=8)

    # Set the title and axis labels with larger font size
    plt.title('Loss Comparison: File 1 vs File 2', fontsize=20)
    plt.xlabel('Epoch', fontsize=16)
    plt.ylabel('Loss', fontsize=16)

    # Customize the tick parameters for larger numbers on the axes
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Add a legend and grid
    plt.legend(fontsize=14)
    plt.grid(True)

    # Annotate the points with loss values
    for i, loss in enumerate(losses_1):
        plt.annotate(f'{loss:.4f}', (i+1, loss), textcoords="offset points", xytext=(0,10), ha='center', fontsize=12)

    for i, loss in enumerate(losses_2):
        plt.annotate(f'{loss:.4f}', (i+1, loss), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=12)

    # Adjust layout and save the figure
    plt.tight_layout()
    plt.savefig(output_filename)
    plt.show()

    print(f"The graph has been saved as '{output_filename}'")


# Usage example
plot_loss_comparison('epoch_losses40T8x8.pkl', 'epoch_losses10T8x8.pkl', 'loss_comparison_40T_10T.png')
```

```python
# Carica le loss dal primo file
with open('epoch_losses20T16x16.pkl', 'rb') as f:
    losses_40T = pickle.load(f)

# Carica le loss dal secondo file
with open('epoch_losses10T16x16.pkl', 'rb') as f:
    losses_10T = pickle.load(f)

# Crea il grafico
plt.figure(figsize=(12, 6))

# Plotta le loss per 40T
plt.plot(range(1, len(losses_40T) + 1), losses_40T, marker='o', linestyle='-', label='20T')

# Plotta le loss per 10T
plt.plot(range(1, len(losses_10T) + 1), losses_10T, marker='s', linestyle='-', label='10T')

plt.title('Confronto Loss: 20T vs 10T')
plt.xlabel('Epoca')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Aggiungi i valori delle loss sopra ogni punto
for i, loss in enumerate(losses_40T):
    plt.annotate(f'{loss:.4f}', (i+1, loss), textcoords="offset points", xytext=(0,10), ha='center')

for i, loss in enumerate(losses_10T):
    plt.annotate(f'{loss:.4f}', (i+1, loss), textcoords="offset points", xytext=(0,-15), ha='center')

plt.tight_layout()
plt.savefig('loss_comparison_40T_10T.png')
plt.show()

print("Il grafico  stato salvato come 'loss_comparison_40T_10T.png'")
```
